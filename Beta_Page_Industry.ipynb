```python
# Importing the important libraries
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
```

```python
# Data time period
years = 5

today = dt.date.today() # date for today
start_date = today - dt.timedelta(days=365 * years) # current date minus years
```

```python
# fetch the data for page industry
page_indus = yf.download('PAGEIND.NS', start=start_date, end=today, interval='1mo')

# Fetch the market datasets as nifty 50
nifty50 = yf.download('^NSEI', start=start_date, end=today, interval='1mo')
```

```python
page_indus
```

```python
nifty50
```

```python
# Calculate monthly returns of page industry
page_indus["Page_Return"] = page_indus['Adj Close'].pct_change().dropna()
page_indus = page_indus.drop(["Open", "Close", "Low", "High","Volume"], axis=1).dropna() # delete the unwanted columns

#Calculate the monthly return for nifty 50
nifty50["Nifty_Return"] = nifty50['Adj Close'].pct_change().dropna()
nifty50 = nifty50.drop(["Open", "Close", "Low", "High","Volume"], axis=1).dropna() # delete the unwanted columns
```

```python
# Checking the columns
page_indus.columns
```

```python
# checking the columns
nifty50.columns
```

```python
# merge in one dataframe
df = pd.merge(page_indus, nifty50, on="Date")
```

```python
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
sns.regplot(x='Page_Return', y='Nifty_Return', data=df)
```

```python
df
```

```python
# Calculate the covariance between the two returns
covariance = np.cov(df['Page_Return'], df['Nifty_Return'])[0][1]
covariance
```

```python
# Calculate the variance of the market returns
market_variance = np.var(df['Nifty_Return'])
market_variance
```

```python
# Calculate the beta
beta = round(covariance / market_variance,3)

# Print the beta
print(fr"The Beta for Page Industry Limited as compare to Nifty 50 is: {beta}")
```
