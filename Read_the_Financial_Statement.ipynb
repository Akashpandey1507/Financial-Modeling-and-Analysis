# Reading the Financial Statement using Python

```python

import yfinance as yf

#import libraries
```

# Balance Sheet

```python
# Create the functions forget access for balance sheet
def get_balance_sheet_data(symbol):
    # for company symbol
    company = yf.Ticker(symbol)
    # get access of balancesheet
    balance_sheet = company.balance_sheet
    return balance_sheet

symbol = "TCS"
# balance sheet
balance_sheet_data = get_balance_sheet_data(symbol)
balance_sheet_data
```

# Cash Flow Statement

```python
# Create the function for cash Flow statement
def get_cash_flow_data(symbol):
    company = yf.Ticker(symbol)
    cash = company.cash_flow
    return cash

symbol = "TCS" # Change the comapny symbol as per your requirement
get_cash_flow_data = get_cash_flow_data(symbol)
get_cash_flow_data
```

# Income Statement

```python

# Create the function for Income Statement
def get_income_statement_data(symbol):
    company = yf.Ticker(symbol)
    inc = company.financials
    return inc

symbol = "TCS" # Change the comapny symbol as per your requirement
get_income_statement_data = get_income_statement_data(symbol)
get_income_statement_data
```
